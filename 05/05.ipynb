Statistical tests scipy: https://docs.scipy.org/doc/scipy/reference/stats.html#statistical-tests

F test not in python, calculate it manually.

```python
a = df_men["IQ"].values
b = df_women["IQ"].values
n = len(a)
m = len(b)

def get_p_both(p_left):
    p_both = 2 * min(p_left, 1-p_left)
    return p_both 

stat = np.var(a, ddof=1) / np.var(b, ddof=1)
p_left = stats.f.cdf(x=stat, dfn=n-1, dfd=m-1)
p_both = get_p_both(p_left)
p_both
```
